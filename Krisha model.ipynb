#### Data description

Data scrapped from Krisha.kz contains 100 records and 9 columns from the first 5 searched pages. 
+ id: a unique number of each apartment

+ title: the title of the apartment in the website

+ price: the price of the selling apartment

+ city: the city where apartment is located 
        
+ number: the contacting phone number of the seller

+ owner: company or individual owner of the apartment or posting the advertisement

+ district: district or region where apartment is located

+ square: area of the apartment, in m²(metres squared)

+ address: physical address, street or location of the apartment

```python
#importing libraries
import pandas as pd
import numpy as np
import re

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel (r'aparts2.xlsx',index_col='Unnamed: 0')
df.head()
```

```python
df['room']=df.title.str[0]
df.head()
```

```python
df['room']=df['room'].astype(int)
df['price']=df['price'].str.strip(' 〒')
df['cost'] = df['price'].str.replace('\xa0', '')

df['cost']=df['cost'].astype(str).astype(int)

df['square']=df['square'].str.strip(' м²')
df['square']=df['square'].astype(float)
```

```python
df.head()
```

```python
df_without = df
df['price'] = [ 'more 30M' if x > 30000000
                  else 'less'
                  for x in df['cost']]

```

```python
df['price'] = [0 if x == 'less' else 1 for x in df['price']]
```

```python
y = df['price']
df=df.drop(['cost','price'],1)
y
```

```python
df['district_cat'] = [ 'Нурсая' if re.compile(r'Нурсая').search(x)
                  else 'Жилгородок' if re.compile(r'Жилгородок').search(x)
                  else 'Алмагуль' if re.compile(r'Алмагуль').search(x)
                  else 'Авангард' if re.compile(r'Авангард').search(x)
                  else 'Центральный'if re.compile(r'Центральный').search(x)
                  else 'Самал' if re.compile(r'Самал').search(x)
                  else 'Привокзальный'if re.compile(r'Привокзальный').search(x)
                  else 'Ардагер' if re.compile(r'Ардагер').search(x)
                  else 'Балыкши' if re.compile(r'Балыкши').search(x)
                  else 'Other'
                  for x in df['title']]
print(df['district_cat'].value_counts().sort_values(ascending=False))
```

```python
df['coast'] = [ 'europe' if re.compile('Абулхаир|Абылхаир|Жарбосынова|Береговая|Нурсая|Сатпаева|Владимирского|Ардагер|Ерниязова|Студенческий|Лесхоз|Самал|Тайманова|Курмангазы|Жарбосынова|Авангард|Кулманов').search(x)
                  else 'asia' if re.compile('Жилгородок|Датов|Чехова|Валиханова|Алмагуль|Азаттык|Атамбаева|Центральный|Балыкши|Привокзальный|Баймуханов|Утемисов|Азаттық').search(x)
                  else 'Other'
                  for x in df['title']]
print(df['coast'].value_counts().sort_values(ascending=False))
```

```python
dummy_column=['coast','district_cat','owner_cat']
```

```python
def toDummy(df, dummy_column):
    for x in dummy_column:
        dummies=pd.get_dummies(df[x],prefix=x,dummy_na=False)
        df = df.drop(x,1)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
df = toDummy(df, dummy_column)
print(df.head(5))
```

### Dropping unnessary columns

```python
df = df.drop(['title','addresss','city','number','owner','district'], axis = 1)
print(df.head(5))
```

### Creating combinations of values

```python
from itertools import combinations
from sklearn.preprocessing import PolynomialFeatures

combos = list(combinations(list(df.columns),2))
colnames = list(df.columns)+['_'.join(x) for x in combos]
poly = PolynomialFeatures(interaction_only=True, include_bias=False)
df=poly.fit_transform(df)
df=pd.DataFrame(df)
df.columns=colnames
noint_indices = [i for i, x in enumerate(list((df==0).all())) if x]
df=df.drop(df.columns[noint_indices],axis=1)

print(df.head())
```

### Normalizing

```python
from sklearn import preprocessing

normalized = pd.DataFrame(preprocessing.normalize(df.square.values.reshape(1,-1)))

normalized=normalized.transpose()
df['square_norm']= normalized
df=df.drop('square',axis=1)
df
```

```python
X =df.drop('Unnamed: 0',1)
```

### Feature selection

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.70, random_state=1)
```

```python
import sklearn.feature_selection

select = sklearn.feature_selection.SelectKBest(k=20)
selected_features = select.fit(X_train, y_train)
indices_selected = selected_features.get_support(indices=True)
colnames_selected = [X.columns[i] for i in indices_selected]

X_train_selected = X_train[colnames_selected]
X_test_selected = X_test[colnames_selected]
```

```python
print(colnames_selected)
```

### Building Logistic Regression model with prepocessed data 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

def find_model_perf(X_train, y_train, X_test, y_test):
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_hat = [x[1] for x in model.predict_proba(X_test)]
    auc = roc_auc_score(y_test, y_hat)
    
    return auc
```

```python
auc_processed = find_model_perf(X_train_selected, y_train, X_test_selected, y_test)
print(auc_processed)
```
